First, let's import the necessary classes from the `transformers` library: `pipeline` for high-level usage, or `AutoTokenizer` and `AutoModelForSequenceClassification` for more control.

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
```

### Option 1: Using `pipeline` for quick and easy sentiment analysis

The `pipeline` function abstracts away much of the complexity, making it easy to perform common tasks like sentiment analysis with just a few lines of code.

```python
from transformers import pipeline

# Load a sentiment-analysis pipeline
sentiment_analyzer = pipeline('sentiment-analysis')

# Analyze some text
result = sentiment_analyzer('I love using Hugging Face Transformers!')
print(result)

result = sentiment_analyzer('This is a good movie but less hardwork dedication i dont like that much movie.')
print(result)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Specify the pre-trained model to use (e.g., a BERT-based model fine-tuned for sentiment analysis)
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the model
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Define the text to analyze
text = "The Transformers library is incredibly not powerful and noteasy to use!"

# Tokenize the text (encode)
inputs = tokenizer(text, return_tensors="pt")
print("Tokens:",inputs)

# Extract input_ids for decoding
encoded_ids = inputs['input_ids'][0]
print(f"Encoded IDs: {encoded_ids.tolist()}")

# Decode the token IDs back to text
decoded_text = tokenizer.decode(encoded_ids)
print(f"Decoded Text: '{decoded_text}'")

# Perform inference
outputs = model(**inputs)

# Get the predicted class (logits)
logits = outputs.logits

# Convert logits to probabilities (optional, but good for understanding)
import torch.nn.functional as F
probabilities = F.softmax(logits, dim=1)

# Get the predicted label
predicted_class_id = logits.argmax().item()
predicted_label = model.config.id2label[predicted_class_id]

print(f"Text: '{text}'")
print(f"Predicted Label: {predicted_label}")
print(f"Probabilities: {probabilities.tolist()}")
```

# **Dataset**

```python
import kagglehub
from datasets import load_dataset
import os

# Download the dataset from Kaggle
# This will return the path to the directory where the dataset files are located.
download_dir_path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")
print(f"Dataset downloaded to: {download_dir_path}")

# The 'tmdb/tmdb-movie-metadata' dataset typically contains CSV files like 'tmdb_5000_movies.csv'.
# You might need to inspect the `download_dir_path` to find the exact filenames.

# Construct the full path to the CSV file you want to load
# This assumes 'tmdb_5000_movies.csv' is directly within the downloaded directory.
csv_file_path = os.path.join(download_dir_path, "tmdb_5000_movies.csv")

# Load the dataset using the 'csv' builder from the datasets library
try:
    dataset = load_dataset('csv', data_files=csv_file_path)
    print("Dataset loaded successfully:")
    print(dataset)
except FileNotFoundError:
    print(f"Error: The file '{csv_file_path}' was not found after download.")
    print("Please check the heading of first page content the downloaded directory: ")
    for root, dirs, files in os.walk(download_dir_path):
        for f in files:
            print(os.path.join(root, f))
except Exception as e:
    print(f"An error occurred while loading the dataset: {e}")
```
