First, let's import the necessary classes from the `transformers` library: `pipeline` for high-level usage, or `AutoTokenizer` and `AutoModelForSequenceClassification` for more control.

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
```

### Option 1: Using `pipeline` for quick and easy sentiment analysis

The `pipeline` function abstracts away much of the complexity, making it easy to perform common tasks like sentiment analysis with just a few lines of code.

```python
from transformers import pipeline

# Load a sentiment-analysis pipeline
sentiment_analyzer = pipeline('sentiment-analysis')

# Analyze some text
result = sentiment_analyzer('I love using Hugging Face Transformers!')
print(result)

result = sentiment_analyzer('This is a good movie but less hardwork dedication i dont like that much movie.')
print(result)
```

### Option 2: Using `AutoTokenizer` and `AutoModel` for more control

If you need more fine-grained control over the tokenization and model inference process, you can load the tokenizer and model separately. This is useful for tasks like custom dataset preparation or integrating with other ML frameworks.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Specify the pre-trained model to use (e.g., a BERT-based model fine-tuned for sentiment analysis)
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the model
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Define the text to analyze
text = "The Transformers library is incredibly not powerful and noteasy to use!"

# Tokenize the text
inputs = tokenizer(text, return_tensors="pt")

# Perform inference
outputs = model(**inputs)

# Get the predicted class (logits)
logits = outputs.logits

# Convert logits to probabilities (optional, but good for understanding)
import torch.nn.functional as F
probabilities = F.softmax(logits, dim=1)

# Get the predicted label
predicted_class_id = logits.argmax().item()
predicted_label = model.config.id2label[predicted_class_id]

print(f"Text: '{text}'")
print(f"Predicted Label: {predicted_label}")
print(f"Probabilities: {probabilities.tolist()}")
```
