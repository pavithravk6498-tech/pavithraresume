Now, we can import the `pipeline` function from `transformers` and load a pre-trained summarization model. We'll use the 'sshleifer/distilbart-cnn-12-6' model, which is a good general-purpose summarization model.

#use a pipeline as a high level  helper

Now, let's define a sample text and summarize it using the loaded model. You can replace this sample text with any document you wish to summarize.

```python
import kagglehub
path = kagglehub.dataset_download("algozee/ai-generated-vs-human-written-text-dataset")
```

```python
from transformers import pipeline
text_generator = pipeline(
    "text-generation",
    model="distilgpt2",
    truncation=True,
    num_return_sequences=2
)
# The original print statement was problematic and 'generated_text' is not defined here.
# You would typically call text_generator with a prompt to get 'generated_text',
# for example: generated_text = text_generator("Some input text")
```

```python
# Define a prompt for text generation
prompt_text = "The quick brown fox jumps over the lazy dog because"

# Generate text based on the prompt
generated_text = text_generator(prompt_text, max_length=15, num_return_sequences=2)

# Print the original prompt and the generated texts
print("Prompt:", prompt_text)
print("\nGenerated Texts:")
for i, generated_sequence in enumerate(generated_text):
    print(f"Sequence {i+1}: {generated_sequence['generated_text']}")
```
